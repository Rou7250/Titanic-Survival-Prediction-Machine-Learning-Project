```python
def social_status(row):
    if row['contarcy_title'] in ['Miss','Mrs', 'Mr','Master','Ms','Mlle', 'Mme']:
        return 'Common'
    elif row['contarcy_title'] in ['Don', 'Dona', 'the Countess','Sir','Lady','Jonkheer']:
        return 'Royal' 
    if row['contarcy_title'] in ['Capt','Major', 'Col']:
        return 'Military'
    else:
        return'Other Profession'
```

```python
df['social_status']=df.apply(social_status,axis=1)
```

```python
df
```

```python
df.loc[df['fam_cnt']==0]
```

```python
tick=df.loc[(df['fam_cnt']+1!=df['psg_cnt']),'ticket'].unique()
```

```python
def comp_type(row):
    if row['psg_cnt']==1:
        return 'Solo'
    else:
        if row['fam_cnt']+1==row['psg_cnt'] or (row['fam_cnt']>=row['psg_cnt']):
            return 'Family'
        elif row['ticket'] in tick:
            return 'Family  + Friends'
        else:
            return 'Friends'   
        
df['companion_type']=df.apply(comp_type,axis=1)
```

# DATA ANALYSIS USING VISUALS

*AIM*: To understand whtich factors are affecting the survival of passengers 

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(7,4))
sns.countplot(df,x='gender',ax=axes[0])
sns.barplot(df,x='gender',y='survived',estimator='mean',ax=axes[1])
plt.suptitle('Gender wise survival analysis')
plt.tight_layout()
plt.show()
```

*Observation:* 

1.Male passengers are more compared to female psgs.

2.Female passengers have high chance of survival than male psgs

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(6,3))
sns.barplot(df,x='gender',y='fare',estimator='sum',ax=axes[0])
sns.barplot(df,x='gender',y='fare',estimator='mean',ax=axes[1])
axes[0].set_ylabel('sum of fare')
axes[1].set_ylabel('ave of fare')
plt.suptitle('Gender wise fare analysis')
plt.tight_layout()
plt.show()
```

**Observation:**

1.Male psgs total fare is higher compared to female.But avg fare is higher in female pags than Male. 

2.This shows that every one are not paying same amount

```python
fig,axes=plt.subplots(ncols=1,nrows=3,figsize=(5,9))
sns.countplot(df,x='pclass',hue='gender',ax=axes[0])
sns.barplot(df,x='pclass',hue='gender',y='fare',estimator='sum',ax=axes[1])
sns.barplot(df,x='pclass',hue='gender',y='fare',estimator='mean',ax=axes[2])
axes[0].set_ylabel('count of fare')
axes[1].set_ylabel('sum of fare')
axes[2].set_ylabel('average of fare')
plt.suptitle('gender and pclass wise fare analysis')
plt.tight_layout()
plt.show()
```

**Observation**:

1.Male psgs are more in number compared to female,hence the total amount is higher in male category.

2.Female avg fare in each pclass is either higher or almost equal to male psgs.
